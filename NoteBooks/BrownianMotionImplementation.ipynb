```python
import math
import sys
import matplotlib.pyplot as plt
import numpy as np

np.set_printoptions(precision=5)
np.random.seed(10)
NbScenarios = 100
k = 50
M = np.zeros((k + 1, NbScenarios + 1))
Mk = np.zeros(k + 1)

for omega in range(1, NbScenarios + 1):
    M[0][omega] = 0
    for j in range(1, k + 1):
        U = np.random.uniform()
        if U < 1/2:
            X = -1
        else:
            X = 1
        M[j][omega] = M[j - 1][omega] + X  # Recursive calculation

Mk = M[k, :]
plt.hist(Mk, bins=15)
plt.show()
print('mean Mk', np.mean(Mk), 'for k=', k)
print('var Mk', np.var(Mk), 'for k=', k)

# Brownian motion
n = 2
Wn = np.sqrt(1/n) * M  # Element-wise scaling

# Histogram of Brownian motion at time t
t = 5
Wn_at_t = Wn[t, :]  # Extract values at time t
plt.hist(Wn_at_t, bins=15)  # Create histogram
plt.title(f'Brownian Motion at t={t}')  # Add title
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.set_printoptions(precision=5)
np.random.seed(10)
NbScenarios = 100
k = 50
n = 2  # Speed of Brownian motion
t = 5   # Time for Brownian motion
N = 1000 # Accuracy parameter

# Initialize arrays
Wn = np.zeros((k + 1, NbScenarios + 1))

# Generate Brownian motion directly
for omega in range(1, NbScenarios + 1):
    Wn[0, omega] = 0  # Initialize Wn at time 0
    for j in range(1, k + 1):
        U = np.random.normal() / N  # Replace uniform with normal
        Wn[j, omega] = Wn[j - 1, omega] + U  # Update Wn recursively

# Plot the histogram
Wn_at_t = Wn[t, :]
plt.hist(Wn_at_t, bins=15)
plt.title(f'Brownian Motion at t={t}, n={n}')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
r = 0.05  # Risk-free rate
sigma = 0.20  # Volatility
S0 = 100  # Initial stock price
T = 5  # Time horizon
NbScenarios = 5000  # Number of scenarios

# Time step (you can adjust this for finer granularity)
dt = 1/252  # Assuming 252 trading days in a year

# Generate random numbers
Z = np.random.standard_normal((NbScenarios, int(T / dt)))

# Simulate stock prices using GBM formula
S = np.zeros_like(Z)
S[:, 0] = S0  # Initialize with initial stock price

for t in range(1, int(T / dt)):
    S[:, t] = S[:, t - 1] * np.exp((r - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z[:, t - 1])

# Sample Mean
sample_mean = np.mean(S, axis=0)  # Mean across scenarios for each time step

# Sample Expected Return
sample_expected_return = (S[:, -1] - S0) / S0  # Return for each scenario

# Histogram of Expected Returns
plt.hist(sample_expected_return, bins=15)
plt.title('Histogram of Sample Expected Returns')
plt.xlabel('Expected Return')
plt.ylabel('Frequency')
plt.show()

# Plot of 10 Scenarios
plt.plot(S[:10, :].T)  # Plot the first 10 scenarios
plt.title('Stock Price Simulation (10 Scenarios)')
plt.xlabel('Time Step')
plt.ylabel('Stock Price')
plt.show()

# Plot of Sample Mean with Expected Value
time_steps = np.arange(S.shape[1])  # Time steps for x-axis
expected_value = S0 * np.exp(r * time_steps * dt)  # Expected value at each time step
plt.plot(time_steps, sample_mean, label='Sample Mean')
plt.plot(time_steps, expected_value, label='Expected Value', linestyle='--')
plt.title('Sample Mean and Expected Value')
plt.xlabel('Time Step')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```
