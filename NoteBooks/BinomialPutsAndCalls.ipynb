```python
import math
import numpy as np

# Parameters
S0 = 100  # Initial stock price
K = 100  # Strike price
sigma = 0.20  # Volatility
dt = 1.0  # Time step
r = 0 # Risk-free interest rate
divyield = 0  # Dividend yield
n = 3  # Number of time steps

# Calculate up and down factors
u = math.exp(sigma * math.sqrt(dt))  # Up factor
d = 1 / u  # Down factor

# Risk-neutral probability
p = (math.exp((r - divyield) * dt) - d) / (u - d)

# Initialize price matrices for European and American options
EurCallPrice = np.zeros((n + 1, n + 1))
USCallPrice = np.zeros((n + 1, n + 1))
EurPutPrice = np.zeros((n + 1, n + 1))
USPutPrice = np.zeros((n + 1, n + 1))

# Print parameters
print(f"u: {u}")
print(f"d: {d}")
print(f"p: {p}")

# Function to calculate option prices using the binomial tree
def calculate_option_prices(n, S0, K, u, d, p, r, dt):
    for t in range(n, -1, -1):  # Iterate backward through time steps
        for i in range(t + 1):  # Iterate through nodes at each time step
            S = S0 * (u ** i) * (d ** (t - i))  # Stock price at node (t, i)
            CallEx = max(S - K, 0)  # Call exercise value
            PutEx = max(K - S, 0)  # Put exercise value

            if t == n:  # At maturity, option price = exercise value
                EurCallPrice[t][i] = CallEx
                USCallPrice[t][i] = CallEx
                EurPutPrice[t][i] = PutEx
                USPutPrice[t][i] = PutEx
            else:  # Before maturity, calculate discounted expected value
                EurCallPrice[t][i] = np.exp(-r * dt) * (EurCallPrice[t + 1][i + 1] * p + EurCallPrice[t + 1][i] * (1 - p)) # Fixed parenthesis here
                USCallCont = np.exp(-r * dt) * (
                    USCallPrice[t + 1][i + 1] * p + USCallPrice[t + 1][i] * (1 - p))
                USCallPrice[t][i] = max(USCallCont, CallEx)  # American call can be exercised early

                EurPutPrice[t][i] = np.exp(-r * dt) * (
                    EurPutPrice[t + 1][i + 1] * p + EurPutPrice[t + 1][i] * (1 - p)) # Fixed parenthesis here
                USPutCont = np.exp(-r * dt) * (
                    USPutPrice[t + 1][i + 1] * p + USPutPrice[t + 1][i] * (1 - p))
                USPutPrice[t][i] = max(USPutCont, PutEx)  # American put can be exercised early

# Calculate option prices
calculate_option_prices(n, S0, K, u, d, p, r, dt)

# Print final option prices
print(f"European Call Option Price: {EurCallPrice[0][0]:.4f}")
print(f"American Call Option Price: {USCallPrice[0][0]:.4f}")
print(f"European Put Option Price: {EurPutPrice[0][0]:.4f}")
print(f"American Put Option Price: {USPutPrice[0][0]:.4f}")
```
